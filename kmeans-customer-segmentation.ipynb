```python
#K-Means Clustering Customer Segmentation
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
df = pd.read_csv("Mall_Customers.csv")
X = df[["Annual Income (k$)", "Spending Score (1-100)"]]
print("------------------------------------------------------------")
print(df.head())
print("------------------------------------------------------------")
print("\n")
wcss = []
for k in [3, 4, 5]:
    kmeans = KMeans(n_clusters=k, random_state=42)
    y_kmeans = kmeans.fit_predict(X)
    #Plotting
    plt.figure(figsize=(6,4))
    for i in range(k):
        plt.scatter(X.values[y_kmeans == i, 0], X.values[y_kmeans == i, 1], 
        label=f"Cluster {i+1}")
    plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:, 1],
        s=100, c='black', label='Centroids')
plt.title(f"K = {k}, WCSS = {kmeans.inertia_:.2f}")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.grid(True)
plt.show()
# Print WCSS
print(f"WCSS for k={k}: {kmeans.inertia_:.2f}")
wcss.append(kmeans.inertia_)
print("\n")
```

```python
k_values = [3, 4,5]
min_wcss = min(wcss)
best_k = k_values[wcss.index(min_wcss)]
print(f"\n Minimm WCSS is (min_wcss:.2f) for k = {best_k}")
```
