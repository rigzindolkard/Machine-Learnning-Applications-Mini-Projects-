```python
# A program to predicting the price of a house using multiple regression. Use the California Housing Dataset from the scikit-learn library. Refer to the link provided. Print MSE and R2 of the model.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the California Housing dataset
data = fetch_california_housing()
x = pd.DataFrame(data.data, columns = data.feature_names)
y = data.target # Target variable (median house value in $100, 000)

```

```python
# split the dataset into training and testing set
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# train the multiple regression model
model = LinearRegression()
model.fit(x_train, y_train)

LinearRegression()

# predict the price of house
y_pred = model.predict(x_test)

# evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# print results
print(f"predicted house prices:{y_pred[:5]}")
print(f"Mean Squared Error (MSE) : {mse:.4f}")
print(f"R-squared (R2) Score: {r2: .4f}")
```

```python
# plot actual vs predicted values
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha = 0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted house prices")
plt.show()
```
